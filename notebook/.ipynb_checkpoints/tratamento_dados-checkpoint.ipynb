# SISTEMA DE RECOMENDAÇÃO DE LIVROS
Fonte do dataset: https://www.kaggle.com/datasets/zygmunt/goodbooks-10k

```python
import pandas as pd
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
livros_df = pd.read_csv("books.csv")
livros_df.head()
```

```python
livros_df = livros_df[["id", "original_title", "ratings_count"]]
livros_df.head()
```

```python
livros_df = livros_df.rename(columns={"id": "ID_LIVRO", "original_title":"TITULO_LIVRO", "ratings_count": "QTD_AVALIACOES"})
livros_df.head()
```

```python
livros_df.shape
```

```python
livros_df = livros_df[livros_df["QTD_AVALIACOES"] > 999]
livros_df.shape
```

```python
livros_df.isna().sum()
```

```python
livros_df = livros_df.dropna()
```

```python
livros_df.isna().sum()
```

```python
avaliacoes_df = pd.read_csv("ratings.csv")
avaliacoes_df.head()
```

```python
avaliacoes_df = avaliacoes_df.rename(columns={"book_id": "ID_LIVRO", "user_id": "ID_USUARIO", "rating": "AVALIACAO"})
avaliacoes_df.head()
```

```python
avaliacoes_df.shape
```

```python
avaliacoes_df.isna().sum()
```

```python
avaliacoes_df["ID_USUARIO"].value_counts()
```

```python
qtd_avaliacoes = avaliacoes_df["ID_USUARIO"].value_counts() > 99
x = qtd_avaliacoes[qtd_avaliacoes].index
x.shape
```

```python
avaliacoes_df.shape
```

```python
avaliacoes_df = avaliacoes_df[avaliacoes_df["ID_USUARIO"].isin(x)]
avaliacoes_df.shape
```

```python
livros_df.info()
```

```python
avaliacoes_df.info()
```

```python
livros_avaliacoes_df = avaliacoes_df.merge(livros_df, on="ID_LIVRO")
livros_avaliacoes_df.head()
```

```python
livros_avaliacoes_df.shape
```

```python
livros_avaliacoes_df = livros_avaliacoes_df.drop_duplicates(["ID_USUARIO", "ID_LIVRO"])
livros_avaliacoes_df.shape
```

```python
livros_pivot = livros_avaliacoes_df.pivot_table(columns="ID_USUARIO", index="TITULO_LIVRO", values="AVALIACAO")
livros_pivot.head()
```

```python
livros_pivot = livros_pivot.fillna(0)
livros_pivot.head()
```

```python
livros_sparse = csr_matrix(livros_pivot)
type(livros_sparse)
```

```python
modelo = NearestNeighbors(algorithm="brute", metric="cosine")
modelo.fit(livros_sparse)
```

```python
titulo_livro = "The Hunger Games"
try:
    index = livros_pivot.index.get_loc(titulo_livro)
except KeyError:
    print("Livro não encontrado na base de dados!")
    index = None
    
if index is not None:
    distances, suggestions = modelo.kneighbors(livros_pivot.values[index].reshape(1, -1))
    print(f"Recomendações para {titulo_livro}:")
    for i in suggestions[0]:
        print(livros_pivot.index[i])
```
